## **1. Contexto Dataset**

### **Descripción de columnas - Dataset Cierres Cartera**


##### Identificadores y datos de la operación
- *NUMERO DE LA COLOCACION2*: Código o consecutivo único del crédito o préstamo.
- *NUMERO DE IDENTIFICACION DEUDOR*: Documento del cliente titular del crédito.
- *NUMERO DEL PRODUCTO*: Código del tipo de producto financiero otorgado.
- *NUMERO DE LA AGENCIA*: Código de la sucursal o agencia donde se gestionó el crédito.

---

##### Información del crédito
- *PLAZO UNIDAD TIEMPO*: Número de cuotas o duración del crédito en una unidad de tiempo (ej. meses).
- *PLAZO EN DIAS*: Plazo total del crédito expresado en días.
- *NUMERO DE PAGOS PACTA*: Cantidad de pagos acordados al inicio del crédito.
- *VALOR ORIGINAL DEL PR**: Monto del préstamo otorgado inicialmente.
- *CAPITAL PEND POR FACT*: Valor del capital aún pendiente por facturar.
- *SALDO DEL PRESTAMO*: Monto actual pendiente de pago del préstamo.

---

##### Fechas importantes
- *FECHA APERTURA DEL PR*: Fecha de apertura o formalización del crédito.
- *FECHA DE DESEMBOLS*: Fecha en la que se hizo efectivo el desembolso.
- *FEC CUOTA MAS ANTIG PENDIENTE DE PAGO*: Fecha de la cuota más antigua que aún no ha sido pagada.
- *FEC PROX PAGO*: Fecha programada para el siguiente pago.
- *FECHA EFEC ULTIMO PAGO*: Fecha en la que se realizó el último pago.
- *FECHA DE NACIMIENTO TITULAR*: Fecha de nacimiento del cliente.

---

##### Estado de pagos y comportamiento
- *NRO PLAN PAGOS VIGENT*: Cantidad de planes de pago aún vigentes.
- *NRO DE UTILIZACINES D*: Veces que se ha utilizado algún beneficio o línea de crédito (columna poco clara, revisar ortografía).
- *DIAS DE MORA*: Número de días que lleva en mora el cliente.
- *CALIFICACION CARTERA*: Clasificación del crédito según el nivel de riesgo o mora (ej. A, B, C...).
- *VLR ULTIMO PAGO*: Valor del último pago realizado por el cliente.
- *VLR TASA*: Valor o porcentaje de la tasa de interés aplicada.
- *VLR CUOTA MES*: Valor mensual que debe pagar el cliente (cuota pactada).
- *VLR MINIMO DE PAGO*: Monto mínimo que puede pagar para evitar mora.
- *DIA VENCIMIENTO CUOTA*: Día del mes en que vence la cuota.
- *NRO CUOTAS VENCIDAS*: Número total de cuotas vencidas hasta la fecha.
- *NRO CUOTAS PAGADAS*: Cuotas ya pagadas por el cliente.
- *NRO CUOTAS RESTANTE*: Cuotas aún pendientes de pago.
- *VALOR CUOTAS VENCIDAS*: Suma de las cuotas vencidas.
- *VALOR CAPITAL CUOTAS VENCIDAS*: Parte de capital dentro de las cuotas vencidas.

---

##### Datos del cliente
- *ESTADO CIVIL*: Estado civil declarado al momento del estudio.
- *SEXO*: Género del cliente.
- *EDAD DEL CLIENTE*: Edad actual del cliente.
- *ESTADO CIVIL.1*: Posible actualización o duplicado del estado civil.
- *NUMERO DE PERSONAS A CARGO*: Personas que dependen económicamente del cliente.
- *INGRESOS*: Ingreso mensual declarado del cliente.
- *OCUPACIÓN*: Actividad laboral del cliente (comerciante, empleado, etc.).
- *ANTIGÜEDAD LABORAL*: Tiempo (en meses) que lleva el cliente trabajando.
- *NIVEL DE ESTUDIOS*: Nivel educativo alcanzado.
- *TIPO DE VIVIENDA*: Tipo de vivienda (propia, arrendada, familiar).

---

##### Vehículo financiado (si aplica)
- *CIUDAD VITRINA*: Ciudad donde se ofreció o entregó el vehículo.
- *MARCA / MARCA.1*: Marca del vehículo adquirido.
- *LINEA*: Línea o modelo específico del vehículo.
- *CLASE*: Tipo de vehículo (camión, carro, moto, etc.).
- *MODELO*: Año del modelo del vehículo.
- *SERVICIO / TIPO DE SERVICIO*: Uso del vehículo (particular, público).
- *TIPO DE VEHICULO*: Si es nuevo, usado, demo, etc.
- *VALOR COMERCIAL*: Valor de mercado del vehículo financiado.

---

##### Otros datos
- *NUMERO DE SOLICITUD TITULAR*: Código único de la solicitud original del crédito.
- *PERFIL DEL TITULAR*: Perfil o segmento del cliente según la entidad.
- *VALOR SOLICITADO*: Monto solicitado por el cliente al inicio del proceso.
- *ASESOR DE DESEMBOLSO / ASESOR DE VENTAS*: Persona responsable del crédito o venta.
- *TASA DE MORA*: Porcentaje aplicado al saldo vencido en caso de mora.
- *TIPO DE ID*: Tipo de documento (CC, CE, NIT, etc.).
- *TIPO DE SOLICITANTE*: Titular, codeudor, etc.
- *ACIERTA*: Valor asociado a scoring crediticio (evaluación automática).
- *PRODUCTO*: Tipo de producto, por ejemplo: cartera castigada, normal, reestructurada, etc.

##  **2. Analisis univariado**

### **Analisis principales variables numerica**

```python
variables = [
    'NUMERO DE PAGOS PACTA', 'VALOR ORIGINAL DEL PR', 'DIAS DE MORA',
    'NRO CUOTAS VENCIDAS', 'NRO CUOTAS PAGADAS', 'NRO CUOTAS RESTANTE',
    'EDAD DEL CLIENTE', 'VALOR COMERCIAL', 'ANTIGÜEDAD LABORAL', 'INGRESOS'
]
```

#### Frecuencia principales variables numericas

```python
# Histograma con KDE para cada variable
plt.figure(figsize=(20, 20))
for i, col in enumerate(variables):
    plt.subplot(4, 3, i + 1)
    sns.histplot(df_cartera[col].dropna(), kde=True)
    plt.title(col)
    plt.xlabel("")
plt.tight_layout()
plt.show()
```

#### Valor atipicos

```python
outlier_details = {}

for var in variables:
    data = df_cartera[var].dropna()
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = max(Q1 - 1.5 * IQR, 0)  # corregido para evitar negativos
    upper_bound = Q3 + 1.5 * IQR

    lower_outliers = data[data < lower_bound]
    upper_outliers = data[data > upper_bound]

    outlier_details[var] = {
        "Total": len(data),
        "Outliers inferiores": len(lower_outliers),
        "Outliers superiores": len(upper_outliers),
        "% Inferiores": round(len(lower_outliers) / len(data) * 100, 2),
        "% Superiores": round(len(upper_outliers) / len(data) * 100, 2),
        "Limite inferior": lower_bound,
        "Limite superior": upper_bound
    }

outlier_stats = pd.DataFrame(outlier_details).T
print(outlier_stats)
```

```python
# Boxplots para cada variable
plt.figure(figsize=(20, 20))
for i, col in enumerate(variables):
    plt.subplot(4, 3, i + 1)
    sns.boxplot(x=df_cartera[col].dropna())
    plt.title(f'Boxplot - {col}')
plt.tight_layout()
plt.show()
```

##  **3. Analisis Bivariado**

```python
# Variables categóricas a analizar
categorical_vars = [
    'SEXO', 'PERFIL DEL TITULAR', 'CLASE', 'MODELO',
    'NIVEL DE ESTUDIOS', 'TIPO DE SOLICITANTE', 'OCUPACIÓN',
    'TIPO DE VIVIENDA', 'ESTADO CIVIL.1', 'TIPO DE VEHICULO'
]

# Crear gráficos bivariados
plt.figure(figsize=(20, 30))
plot_idx = 1

for col in categorical_vars:
    prop_df = pd.crosstab(df_cartera[col], df_cartera['CREDITO EN MORA'], normalize='index') * 100
    valid_cols = [c for c in ['1', '0'] if c in prop_df.columns]
    
    if len(valid_cols) == 0:
        continue  # Saltar si no hay datos válidos


    ax = plt.subplot(6, 2, plot_idx)
    colors = ['red' if c == '1' else 'green' for c in valid_cols]
    prop_df[valid_cols].plot(kind='bar', stacked=True, ax=ax, color=colors, legend=False)
    
    plt.title(f'% de Mora por categoría de {col}')
    plt.ylabel('%')
    plt.xlabel('')
    plt.xticks(rotation=45)
    plot_idx += 1

plt.tight_layout()
plt.legend(title='CREDITO EN MORA', labels=['En mora', 'Sin mora'], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
